```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
```

```python
np.random.seed(613)
x = np.arange(50) # the amount of value we want on the x axis
y1 = np.random.randint(0,100,50) # first value is min, second is max value , third is how many we want to plot on the y axis. 
y2 = np.random.randint(0,100,50)

fig, ax= plt.subplots(figsize =(5,3)) # we want to have the size of this chart that's 5 inches * 3 inches
ax.plot(x,y1)
ax.plot(x,y2)
```

```python
np.random.seed(613)
x = np.arange(50) # the amount of value we want on the x axis
y1 = np.random.randint(0,100,50) # first value is min, second is max value , third is how many we want to plot on the y axis. 
y2 = np.random.randint(0,100,50)

fig, ax= plt.subplots(figsize =(5,3)) # we want to have the size of this chart that's 5 inches * 3 inches
ax.plot(x,y1, label= "Person1")
ax.plot(x,y2, label= "Person2")
ax.legend(loc= "lower right",
   frameon = True,
   fontsize = 12,
   ncol = 2,
   shadow = True)
```

```python
fig, ax = plt.subplots(figsize = (5,3))
ax.plot(x,y1,label = "Person1")
ax.plot(x,y2,label = "Person2")
ax.legend(loc= "upper left",
          bbox_to_anchor=(1,1.2))  
```

```python
fig, ax = plt.subplots(figsize = (6,3))
ax.plot(x,y1,label = "Person1")
ax.plot(x,y2,label = "Person2")
ax.legend(loc = "lower right")
ax.text(10,98, "This value is important") # shows where the annotation starts.
```

```python
fig, ax = plt.subplots(figsize = (6,3))
ax.scatter(x,y1,label = "Person1",
color = "green")
ax.plot(x,y2,label = "Person2")
ax.legend(loc = "lower right")
ax.text(10,110, "This value is important!",
ha = "center",
color = "red",
size = 20) # shows where the annotation starts.
```

```python
fig, ax = plt.subplots(figsize = (7,4))
ax.axis([0,10,0,10])

ax.text(1,5, ". Data: (1,5)", transform=ax.transData)
ax.text(0.5,0.1, ". Axes: (0.5,0.1)", transform=ax.transAxes)
ax.text(0.2,0.2, ". Figure: (0.2,0.2)", transform=fig.transFigure)
```

```python
fig, ax = plt.subplots(figsize = (6,3))
ax.scatter(x,y1,label = "Person1",
color = "green")
ax.plot(x,y2,label = "Person2")
ax.legend(loc = "lower right")
ax.annotate("This value is important!",  xy = (10,95),
xytext=(20,94),
arrowprops=dict(facecolor= 'black', shrink = 0.05)) # shows where the annotation starts.
```

```python
fig, ax = plt.subplots(figsize = (6,3))
ax.scatter(x,y1,label = "Person1",
color = "green")
ax.plot(x,y2,label = "Person2")
ax.legend(loc = "lower right")
ax.annotate("This value is important!",  xy = (10,95),
xytext=(20,94),
arrowprops=dict(facecolor= 'orange', arrowstyle = "wedge")) # shows where the annotation starts.
```

```python
fig, ax = plt.subplots(figsize = (6,3))
ax.scatter(x,y1,label = "Person1",
color = "green")
ax.plot(x,y2,label = "Person2")
ax.legend(loc = "lower right")

ax.yaxis.set_major_locator(plt.NullLocator()) #locator to remove tics and labels
ax.xaxis.set_major_formatter(plt.NullFormatter())#formatter to remove only labels
```

```python
fig, ax = plt.subplots(figsize = (6,3))
ax.scatter(x,y1,label = "Person1",
color = "green")
ax.plot(x,y2,label = "Person2")
ax.legend(loc = "lower right")

ax.xaxis.set_major_locator(plt.MaxNLocator(3)) #locator to locate 3 max values on x axis.
```

```python
fig, ax = plt.subplots(figsize = (6,3))
ax.scatter(x,y1,label = "Person1",
color = "green")
ax.plot(x,y2,label = "Person2")
ax.legend(loc = "lower right")

ax.xaxis.set_major_locator(plt.MultipleLocator(5)) #locator to locate 3 max values on x axis.
```

```python
fig, ax = plt.subplots(figsize = (6,3))
ax.scatter(x,y1,label = "Person1",
color = "green")
ax.plot(x,y2,label = "Person2")
ax.legend(loc = "lower right")

plt.xticks(rotation=45, ha='right', fontsize = 12, font = "serif", color = "red")
```

```python
plt.style.use('fivethirtyeight')
font1={'family' : 'serif', 'color': 'indigo'}
fig, ax = plt.subplots(figsize = (6,3))
ax.scatter(x,y1,label = "Person1",
color = "green")
ax.plot(x,y2,label = "Person2", color = "red")
ax.legend(loc = "lower right")

ax.xaxis.set_major_locator(plt.MultipleLocator(5))
plt.xticks(rotation=45, ha='right', fontsize = 12, font = "serif", color = "red")
plt.xlabel('Shiny New X axis!', fontsize = 18, fontdict = font1)
#plt.style.available
```
