```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
```

```python
np.random.seed(613)
x1 = np.arange(50)
y1 = np.random.randint(0,100,50) # first value is min, second is max value , third is how many we want to plot on the y axis. 
x2 = np.array(["Luffy", "Zoro", "Nami", "Usopp", "Sanji"])
y2 = np.array([110, 180, 240,99,220])
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, nrows=1,figsize =(8,3)) # we want to have the size of this chart that's 5 inches * 3 inches and the number of rows and columns here.
ax1.scatter(x1,y1, color = "red")
ax2.bar(x2,y2, color= "yellow")

ax1.xaxis.set_major_formatter(plt.NullFormatter())
ax1.yaxis.set_major_locator(plt.MultipleLocator(10))

ax2.yaxis.set_major_locator(plt.MultipleLocator(25))
```

```python
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'],
                                    ['ax2', 'ax3']],
                                    figsize=(7,4))

someaxes["ax1"].scatter(x1,y1, color = "green")
someaxes["ax2"].bar(x2,y2, color = "red")    
someaxes["ax3"].plot(x1,y1, color = "yellow")                                
```

```python
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'],
                                    ['ax2', 'ax3']],
                                    figsize=(7,4))

someaxes["ax1"].scatter(x1,y1, color = "green")
someaxes["ax2"].bar(x2,y2, color = "red")    
someaxes["ax3"].plot(x1,y1, color = "yellow")   

someaxes["ax1"].set_xlabel("A Big label", fontsize=18)
someaxes["ax2"].set_xlabel("Another label", fontsize=18)
someaxes["ax3"].set_xlabel("A Third Label!", fontsize=18)
```

```python
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'],
                                    ['ax2', 'ax3']],
                                    figsize=(7,4), layout = "constrained")

someaxes["ax1"].scatter(x1,y1, color = "green")
someaxes["ax2"].bar(x2,y2, color = "red")    
someaxes["ax3"].plot(x1,y1, color = "yellow")   

someaxes["ax1"].set_xlabel("A Big label", fontsize=18)
someaxes["ax2"].set_xlabel("Another label", fontsize=18)
someaxes["ax3"].set_xlabel("A Third Label!", fontsize=18)
```
