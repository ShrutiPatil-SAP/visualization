```python
#We are importing the libraries here.
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
```

```python
#Defining values for Data visualization.
np.random.seed(613)
x = np.arange(50) # the amount of value we want on the x axis
y = np.random.randint(0,100,50) # first value is min, second is max value , third is how many we want to plot on the y axis. 

fig, ax= plt.subplots(figsize =(5,3)) # we want to have the size of this chart that's 5 inches * 3 inches
ax.scatter(x,y)
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.bar(x,y) # This makes it a bar which is vertical
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.barh(x,y) # This makes it a horizontal bar
```

```python
#Below you will see the histogram plotting 
fig, ax = plt.subplots(figsize=(5,3)) # This will depict a size of 5 inches by 3 inches
ax.hist(y) # This shows that y axis is on histogram plot.
```

```python
fig, ax = plt.subplots(figsize=(5,3))

ax.plot(x,y)

ax.set_title('total growth over time')
ax.set_ylabel('total Growth')
ax.set_xlabel('Years since start')
```

```python
font1 = {'family':'sans-serif', 'color':'red', 'size':20}
font2 = {'family':'monospace', 'color':'orange', 'size':14}

fig, ax = plt.subplots(figsize=(5,3))

ax.plot(x,y)

ax.set_title('total growth over time' , fontdict = font1)
ax.set_ylabel('total Growth' , fontdict = font2)
ax.set_xlabel('Years since start' , fontdict = font2)
```

```python
fig, ax = plt.subplots(figsize=(5,3))

ax.scatter(x,y,
           marker = '*',
           color = 'turquoise')
```

```python
fig, ax = plt.subplots(figsize=(5,3))

ax.plot(x,y,
           marker = '*',
           color = 'blue',
           linestyle = '--',
           linewidth = '5')
```

```python
fig, ax = plt.subplots(figsize=(5,3))

ax.plot(x,y,
           marker = '*',
           color = 'blue',
           linestyle = '--',
           linewidth = '3',
           markeredgecolor = '#FFA500',
           markerfacecolor = '#000000')
```

```python
fig, ax = plt.subplots(figsize=(5,3))

ax.plot(x,y,
           marker = '*',
           color = 'green',
           linestyle = 'dashdot',
           linewidth = '2',
           markeredgecolor = '#FFA500',
           markerfacecolor = '#000000')
ax.grid(axis= 'y',
        color = 'blue',
        linestyle = 'dotted')           
```

```python
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(8,8))

# Big bins
axs[0,0].hist2d(x, y, bins=(50, 50), cmap=plt.cm.jet)
axs[0, 0].set_title('bins = (50, 50)')
 
# Small bins
axs[0,1].hist2d(x, y, bins=(600, 600), cmap=plt.cm.jet)
axs[0, 1].set_title('bins = (600, 600)')
 
# If you do not set the same values for X and Y, the bins won't be a square!
axs[1,0].hist2d(x, y, bins=(600, 30), cmap=plt.cm.jet)
axs[1, 0].set_title('bins = (600, 30)')

# If you do not set the same values for X and Y, the bins won't be a square!
axs[1,1].hist2d(x, y, bins=(30, 600), cmap=plt.cm.jet)
axs[1, 1].set_title('bins = (30, 600)')

plt.show()
```

```python
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(8,8))

# Big bins
axs[0,0].hist2d(x, y, bins=(50, 50), cmap=plt.cm.Reds_r)
axs[0, 0].set_title('cmap=plt.cm.Reds')
 
# Small bins
axs[0,1].hist2d(x, y, bins=(50, 50), cmap=plt.cm.Blues_r)
axs[0, 1].set_title('cmap=plt.cm.Blues')
 
# If you do not set the same values for X and Y, the bins won't be a square!
axs[1,0].hist2d(x, y, bins=(50, 50), cmap=plt.cm.Greens_r)
axs[1, 0].set_title('cmap=plt.cm.Greens')

# If you do not set the same values for X and Y, the bins won't be a square!
axs[1,1].hist2d(x, y, bins=(50, 50), cmap=plt.cm.Greys_r)
axs[1, 1].set_title('cmap=plt.cm.Greys')

plt.show()
```
